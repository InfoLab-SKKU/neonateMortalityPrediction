```python
import pandas as pd 
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

#### Loading dataset 

```python
single_raw_results = pd.read_csv("single_models_results/single_models_results_raw.csv")
single_optimized_results = pd.read_csv("single_models_results/single_models_results_optimized.csv")
```

```python
single_raw_results
```

```python
raw_results = single_raw_results[['classifiers', 'accuracy']]
optimized_results = single_optimized_results[['classifiers', 'accuracy']]
raw_results['result'] = 'raw'
optimized_results['result'] = 'optimized'
single_results = pd.concat([raw_results, optimized_results])
```

```python
single_results
```

```python
sns.set_theme()
sns.set(rc={'figure.figsize':(10,5)})
sns.set_style("whitegrid")
```

```python
ax = sns.barplot(x="classifiers", y="accuracy", hue="result", data=single_results)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")

ax.legend(bbox_to_anchor=(1.0, 1.0))
ax.set_ylim(0, 1)
```
